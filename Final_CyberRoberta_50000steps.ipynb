```python
from transformers import RobertaForMaskedLM

class CustomRobertaWithNoise(nn.Module):
    def __init__(self, config):
        super(CustomRobertaWithNoise, self).__init__()
        self.roberta = RobertaForMaskedLM.from_pretrained('roberta-base', config=config)
        self.config=config
        self.add_noise_to_embeddings()

    def add_noise_to_embeddings(self, noise_factor=0.1):
        embeddings = self.roberta.roberta.embeddings.word_embeddings
        # noise tensor with the same shape as the embeddings
        noise = torch.normal(mean=0, std=1, size=embeddings.weight.size())
        noise *= noise_factor
        # Add noise to the pretrained weights
        embeddings.weight.data = embeddings.weight.data + noise.clone()

    def forward(self, input_ids, attention_mask=None, labels=None):
        # Forward pass through the RoBERTa model
        outputs = self.roberta(input_ids, attention_mask=attention_mask, labels=labels)
        return outputs

config = RobertaConfig(
    vocab_size=50265,
    max_position_embeddings=514,
    num_attention_heads=12,
    num_hidden_layers=6,
    type_vocab_size=1,
)
custom_model = CustomRobertaWithNoise(config)
model=custom_model
print(model)
```

```python
# Load cybersecurity dataset
dataset = LineByLineTextDataset(
    tokenizer=tokenizer,
    file_path="/content/Corpus (1).txt",
    block_size=512,
)

# data collator
data_collator = DataCollatorForLanguageModeling(
    tokenizer=tokenizer, mlm=True, mlm_probability=0.15
)
```

```python
from sklearn.model_selection import train_test_split

total_examples = len(dataset)

#  train-val split ratio
train_ratio = 0.9
val_ratio = 1 - train_ratio

# Split the dataset into train and val sets
train_dataset, val_dataset = train_test_split(dataset, test_size=val_ratio, random_state=42)

print("Number of examples in the train dataset:", len(train_dataset))
print("Number of examples in the val dataset:", len(val_dataset))
```

```python

train_file = "/content/train_dataset.txt"
val_file = "/content/val_dataset.txt"

def write_dataset(dataset, file_path):
    with open(file_path, "w", encoding="utf-8") as file:
        for example in dataset:
            example_str = " ".join([str(value) for value in example.values()])
            file.write(example_str + "\n")

# Write train dataset to file
write_dataset(train_dataset, train_file)
print("Train dataset saved to:", train_file)

# Write test dataset to file
write_dataset(val_dataset, val_file)
print("Val dataset saved to:", val_file)
```

```python
training_args = TrainingArguments(
    output_dir="./roberta-retrained",
    overwrite_output_dir=True,
    num_train_epochs=6,
    per_device_train_batch_size=4,
    prediction_loss_only=True,
    evaluation_strategy="steps",
    save_steps=10000,
    metric_for_best_model="eval_loss",
    greater_is_better=False,  # Lower validation loss is better
    do_eval=True,
)

trainer = Trainer(
    model=model,
    args=training_args,
    data_collator=data_collator,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    callbacks=[LossDifferenceCallback()]
)
```

```python
import torch


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)

# function to perform the fill-mask task
def fill_mask(text, top_k=5):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    inputs = {key: tensor.to(device) for key, tensor in inputs.items()}
    mask_token_index = torch.where(inputs["input_ids"] == tokenizer.mask_token_id)
    logits = model(**inputs).logits
    mask_logits = logits[0, mask_token_index[1], :]
    top_indices = mask_logits.topk(top_k)
    top_predictions = [tokenizer.decode(index) for index in top_indices.indices.tolist()]
    top_scores = top_indices.values.tolist()
    return top_predictions, top_scores


predictions, scores = fill_mask("zombie is the other name for a <mask>.")
for prediction, score in zip(predictions, scores):
    print(f"Prediction: {prediction}, Score: {score}")
```
